Experiment No. 3 – Text Cleaning, Lemmatization, Stop Word Removal, Label Encoding and TF-IDF

```python

# ==========================================
# EXPERIMENT 3
# Text Cleaning, Lemmatization,
# Stop Word Removal, Label Encoding, TF-IDF
# ==========================================

# Install required libraries
!pip install nltk scikit-learn pandas

# ==========================================
# Import Libraries
# ==========================================

import pandas as pd
import re
import pickle
import nltk

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder

# Download NLTK data
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('omw-1.4')

# ==========================================
# Sample Dataset
# ==========================================

data = {
    'Content': [
        "Natural Language Processing is amazing!",
        "Machine Learning is very powerful.",
        "Deep Learning models are useful in AI.",
        "NLP uses text preprocessing techniques."
    ],

    'Category': [
        "NLP",
        "ML",
        "AI",
        "NLP"
    ]
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
print(df)

# ==========================================
# Initialize Lemmatizer and Stopwords
# ==========================================

lemmatizer = WordNetLemmatizer()

stop_words = set(stopwords.words('english'))

# ==========================================
# Text Cleaning Function
# ==========================================

def clean_text(text):

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    # Tokenization
    tokens = word_tokenize(text)

    # Lemmatization
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Remove stopwords
    tokens = [token for token in tokens if token not in stop_words]

    # Join tokens
    clean_text = ' '.join(tokens)

    return clean_text

# ==========================================
# Apply Cleaning
# ==========================================

df['clean_text'] = df['Content'].apply(clean_text)

print("\n===================================")
print("CLEANED TEXT")
print("===================================\n")

print(df[['Content', 'clean_text']])

# ==========================================
# LABEL ENCODING
# ==========================================

label_encoder = LabelEncoder()

df['Encoded_Category'] = label_encoder.fit_transform(df['Category'])

print("\n===================================")
print("LABEL ENCODING")
print("===================================\n")

print(df[['Category', 'Encoded_Category']])

# ==========================================
# TF-IDF REPRESENTATION
# ==========================================

tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(df['clean_text'])

tfidf_dataframe = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out()
)

print("\n===================================")
print("TF-IDF REPRESENTATION")
print("===================================\n")

print(tfidf_dataframe)

# ==========================================
# SAVE OUTPUTS
# ==========================================

with open('processed_data.pickle', 'wb') as f:
    pickle.dump(df, f)

with open('tfidf_vectorizer.pickle', 'wb') as f:
    pickle.dump(tfidf_vectorizer, f)

print("\n===================================")
print("FILES SAVED SUCCESSFULLY")
print("===================================\n")

print("1. processed_data.pickle")
print("2. tfidf_vectorizer.pickle")

# ==========================================
# FINAL DATASET
# ==========================================

print("\n===================================")
print("FINAL DATASET")
print("===================================\n")

print(df)

# ==========================================
# CONCLUSION
# ==========================================

print("\n===================================")
print("CONCLUSION")
print("===================================\n")

print("Successfully performed:")
print("1. Text Cleaning")
print("2. Lemmatization")
print("3. Stop Word Removal")
print("4. Label Encoding")
print("5. TF-IDF Representation")
print("6. Saved Processed Outputs")
```
