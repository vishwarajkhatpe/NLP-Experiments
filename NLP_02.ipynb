```python
# ============================================
# EXPERIMENT 2
# Bag of Words, TF-IDF and Word2Vec
# Google Colab Code
# ============================================

# Install required libraries
!pip install nltk gensim scikit-learn pandas

# ============================================
# Import Libraries
# ============================================

import nltk
import pandas as pd
import numpy as np

from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec

# Download tokenizer data
nltk.download('punkt')

# ============================================
# Sample Sentences
# ============================================

string1 = "Data science is fun and interesting"
string2 = "Data science is fun"
string3 = "science is interesting"

corpus = [string1, string2, string3]

print("Corpus:\n")
for sentence in corpus:
    print(sentence)

# ============================================
# BAG OF WORDS USING COUNT VECTORIZER
# ============================================

print("\n===================================")
print("BAG OF WORDS")
print("===================================\n")

count_vectorizer = CountVectorizer()

bow_matrix = count_vectorizer.fit_transform(corpus)

bow_dataframe = pd.DataFrame(
    bow_matrix.toarray(),
    columns=count_vectorizer.get_feature_names_out()
)

print(bow_dataframe)

# ============================================
# COUNT OCCURRENCE
# ============================================

print("\n===================================")
print("COUNT OCCURRENCE")
print("===================================\n")

doc = " ".join(corpus)

count_vec = CountVectorizer()

count_occurs = count_vec.fit_transform([doc])

count_occur_df = pd.DataFrame(
    (count, word)
    for word, count in zip(
        count_vec.get_feature_names_out(),
        count_occurs.toarray()[0]
    )
)

count_occur_df.columns = ['Word', 'Count']

count_occur_df.sort_values('Count', ascending=False, inplace=True)

print(count_occur_df)

# ============================================
# NORMALIZED COUNT OCCURRENCE
# ============================================

print("\n===================================")
print("NORMALIZED COUNT OCCURRENCE")
print("===================================\n")

norm_count_vec = TfidfVectorizer(
    use_idf=False,
    norm='l2'
)

norm_count_occurs = norm_count_vec.fit_transform([doc])

norm_count_occur_df = pd.DataFrame(
    (count, word)
    for word, count in zip(
        norm_count_vec.get_feature_names_out(),
        norm_count_occurs.toarray()[0]
    )
)

norm_count_occur_df.columns = ['Word', 'Normalized Count']

norm_count_occur_df.sort_values(
    'Normalized Count',
    ascending=False,
    inplace=True
)

print(norm_count_occur_df)

# ============================================
# TF-IDF
# ============================================

print("\n===================================")
print("TF-IDF")
print("===================================\n")

tfidf_vec = TfidfVectorizer()

tfidf_matrix = tfidf_vec.fit_transform(corpus)

tfidf_dataframe = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vec.get_feature_names_out()
)

print(tfidf_dataframe)

# ============================================
# WORD2VEC PREPARATION
# ============================================

print("\n===================================")
print("WORD2VEC")
print("===================================\n")

tokenized_data = []

for sentence in corpus:
    tokens = word_tokenize(sentence.lower())
    tokenized_data.append(tokens)

print("Tokenized Data:\n")
print(tokenized_data)

# ============================================
# CBOW MODEL
# ============================================

print("\n===================================")
print("CBOW MODEL")
print("===================================\n")

cbow_model = Word2Vec(
    sentences=tokenized_data,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

print("Similarity between 'data' and 'science':")
print(cbow_model.wv.similarity('data', 'science'))

# ============================================
# SKIP-GRAM MODEL
# ============================================

print("\n===================================")
print("SKIP-GRAM MODEL")
print("===================================\n")

skipgram_model = Word2Vec(
    sentences=tokenized_data,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=1
)

print("Similarity between 'fun' and 'interesting':")
print(skipgram_model.wv.similarity('fun', 'interesting'))

# ============================================
# WORD VECTOR EXAMPLE
# ============================================

print("\n===================================")
print("WORD VECTOR")
print("===================================\n")

print("Vector representation of word 'data':\n")

print(cbow_model.wv['data'])

# ============================================
# CONCLUSION
# ============================================

print("\n===================================")
print("CONCLUSION")
print("===================================\n")

print("Successfully implemented:")
print("1. Bag of Words")
print("2. Count Occurrence")
print("3. Normalized Count Occurrence")
print("4. TF-IDF")
print("5. Word2Vec using CBOW and Skip-Gram")
```
