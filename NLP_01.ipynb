**Whitespace Tokenization**

```python
from nltk.tokenize import WhitespaceTokenizer

wt = WhitespaceTokenizer()

text = "Welcome to NLP"

tokenized_text = wt.tokenize(text)

print(tokenized_text)
```

**Punctuation** Tokenization

```python
from nltk.tokenize import WordPunctTokenizer

text = "Hello, NLP!"

result = WordPunctTokenizer().tokenize(text)

print(result)
```

Treebank tokenizer

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

print(tokenizer.tokenize("I'm learning NLP"))
```

**Tweet** Tokenizer

```python
from nltk.tokenize import TweetTokenizer

tweet = TweetTokenizer()

text = "#NLP is awesome 😊"

print(tweet.tokenize(text))
```

MWE Tokenizer

```python
from nltk.tokenize import MWETokenizer

tokenizer = MWETokenizer()

tokenizer.add_mwe(('New', 'York'))

print(tokenizer.tokenize("I live in Pune".split()))
```

#Stemming

**Porter Stemmer**

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

words = ["playing", "played", "plays"]

for word in words:
    print(ps.stem(word))
```

**Snowball Stemmer**

```python
from nltk.stem import SnowballStemmer

snow = SnowballStemmer(language='english')

print(snow.stem("generously"))
```

Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running"))
```

```python
#pos tagging
lemmatizer.lemmatize("running", pos='v')
```
